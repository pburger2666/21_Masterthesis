```python
import csv
import matplotlib.pyplot as plt
import numpy as np

filelist=["data/r86.dat"]
labellist=["r86"]
filecount=0

fig, ax = plt.subplots()
for name in filelist:
    X=[]
    Y=[[],[],[]]
    linecount=0
    print(name)

    with open(name) as csvfile:
        reader = csv.reader(csvfile,delimiter=',')
        for row in reader:
            linecount+=1
            if linecount<33:
                continue    
            #print(row[7])
            X.append(float(row[7]))
            Y[0].append(float(row[19]))
            Y[1].append(float(row[20]))
            Y[2].append(float(row[21]))

    ax.plot(X,Y[0], '.-',label="cross 1")
    ax.plot(X,Y[1], '.-',label="cross 2")
    ax.plot(X,Y[2], '.-',label="cross 3")

    filecount+=1

ax.set_ylabel('R / [$\Omega$]')
ax.set_xlabel(r'$I_b$ / [$\mu$A]')
plt.grid()
plt.legend()
plt.show()
```
